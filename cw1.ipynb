### 1. Population dataset

#### Analysing the data and performing data wrangling

```python
print(df_population.head())  # print the 5 first rows to show us a brief section of the dataset

print(df_population.info())  # to get information about the dataset, specifically its columns and the following datatypes

print(df_population.describe())  # gives a summary statistic 
```

```python
df_population
```

### Creating graphs and plots

### 1.1 Total Population Growth 

```python
# Creating a lineplot graph to show the overall population growth across the 3 countries in the last 10 years

ten_years = df_population[(df_population["Year"] >= 2013) & (df_population["Year"] <= 2023)]  # filter data for the last 10 years (2013-2023)

df_total_population = ten_years.groupby(["Year", "Country"])["Total_Population"].sum().reset_index()  # aggregrates total population by year and country

plt.figure(figsize=(10, 5))
sns.lineplot(data=df_total_population, x="Year", y="Total_Population", hue="Country", marker="o")
plt.title("Total Population Growth (2013-2023)")
plt.xlabel("Year")
plt.ylabel("Total Population (millions)")
plt.grid(True)
plt.legend(title="Country", loc="upper right" )
plt.show()
```

**Insights**:
From the above graph as all 3 countries show an upward trend this indicates that there is a positive population growth for all 3 countries. 
UK has the highest population which gradually increases linearly telling us that its a stable but aging population which can be due to lower birht rates and controlled immigration policies. 
Malaysia has a more stable growth as it shows a moderate increase which tells us that it may have higher birth rates compared to UK. 
UAE on the other hand has the smallest population comparitively but has a slightly higher growth rate compared to UK and Malaysia which tells us that more expats have moved here and higher migration taking place. UAE’s population is growing faster in percentage terms, but the total population remains much smaller.
Moreover, since none of the countries show a drastic decline in population, this suggests that there were no major demographic crisis.




### 1.2 Total Working Age Population Growth

The source from which the working age is taken is form the world bank group (https://databank.worldbank.org/metadataglossary/gender-statistics/series/SP.POP.DPND)

```python
# Creating a lineplot graph to show the working age population growth over the last 10 years for the 3 countries

# in this case we will assume 15-64 as working age taking from the world bank group data bank 
df_working_population = ten_years[(ten_years["AgeGrpStart"] >= 15) & (ten_years["AgeGrpStart"] <= 64)] # filter age to be only the working population
df_working_population = df_working_population.groupby(["Year", "Country"])["Total_Population"].sum().reset_index() # aggregate working-age population

plt.figure(figsize=(10, 5))
sns.lineplot(data=df_working_population, x="Year", y="Total_Population", hue="Country", marker="o")
plt.title("Working Age Population Growth (2013-2023)")
plt.xlabel("Year")
plt.ylabel("Working Population (millions)")
plt.grid(True)
plt.legend(title="Country")
plt.show()
```

**Insights**:
Analysing the graph we can tell as all 3 countries have an upward trend that the working population increases over the years. UAE has the highest working population, consistently above 40 million with a steady increase. This could indicate an aging workforce where new employers are trying to keep up with the retirements. Malaysia has a steady growth over the 10 years and one of the reasons for this may be because of the higher birth rates than the UK as more young people enter the workforce. Economically this is good for the country as there will be more productivity with a higher number of people working but could also be a strain on job markets if there aren't enough jobs. UAE has the smallest working population but a faster growth than UK, this may be due to UAE having a workforce more keen to expatriate workers.

### Merging the population and GDP datasets

### GDP per capita vs GDP per working population

```python
# Creating a grouped bar chart
countries = df_merged_GDP_population["Country"].tolist()
gdp_per_capita = df_merged_GDP_population["GDP_per_capita"].tolist()
gdp_per_worker = df_merged_GDP_wpopulation["perWorkerGDP"].tolist()

x = np.arange(len(countries))
width = 0.35  # Width of bars

fig, ax = plt.subplots(figsize=(10, 6))

bars1 = ax.bar(x - width/2, gdp_per_capita, width, label='GDP per Capita', color='red')
bars2 = ax.bar(x + width/2, gdp_per_worker, width, label='GDP per Worker', color='blue')

# Labels
ax.set_xlabel("Country")
ax.set_ylabel("GDP Value (per 10 million)")
ax.set_title("Comparison of GDP per Capita vs GDP per Worker")
ax.set_xticks(x)
ax.set_xticklabels(countries)
ax.legend()


# Show plot
plt.show()
```

**Insights**: By observing the bar graph above we can coonclude that the GDP per worker is significantly higher than GDP per capita. This is predicted as the working population is much smaller than the actual population thus leading to a much higher value of GDP per worker. Here UK has the highest GDP per worker and UAE has the highest GDP per Capita. Overall using GDP per worker as a measure provides a better estimate of economic productivity per employed person than GDP per capita and helps understand labor productivity. However, it may overestimate productivity in countries where only a small percent of the population is employed. To conclude, using GDP per worker alone without other indicators may make it hard to properly know what is happening.
